# Question-1

**(a) Post with most likes and post with most comments**

```python
f = open('inc_insta.csv', 'r')
df = pd.read_csv(f)
likes=df['likes']
comments=df['comments']
for i in range(0, len(likes)):
    if 'k' in str(likes[i]):
        likes[i] = float(likes[i].strip('k'))*1000
        continue
    likes[i] = ''.join(c for c in likes[i] if c.isnumeric())
    likes[i] = int(likes[i])
for i in range(0, len(comments)):
    if 'k' in str(comments[i]):
        comments[i] = float(comments[i].strip('k'))*1000
        continue
    comments[i] = ''.join(c for c in comments[i] if c.isnumeric())
    comments[i] = int(comments[i])
for i in range(0,len(likes)):
    if likes[i]==likes.max():
        a=df['id'][i].strip("[]").strip("'")
print("ID : ", a)
print("Link to the post with most likes: https://www.instagram.com/incindia"+a)
for i in range(0,len(comments)):
    if comments[i]==comments.max():
        print("ID : ", df['id'][i].strip("[]").strip("'"))
        print("Link to the post with most comments: https://www.instagram.com/incindia"+df['id'][i].strip("[]").strip("'"))
```

# Question-2

**(a) Pie chart representing the share of images in the tweets as compared to videos**

```python
fr = open('bjp_insta.csv', 'r')
df = pd.read_csv(fr)

count=0
views = df['views']
for i in range(0, len(views)):
    if 'k' in str(views[i]):
        views[i] = float(views[i].strip('k'))*1000
        continue
    views[i] = ''.join(c for c in views[i] if c.isnumeric())
    views[i] = int(views[i])
for i in df['views']:
    if i == 1:
        count+=1
video=count
photo=len(df)-count
print("No. of images :", photo)
labels = 'Video', 'Photo'
sizes = [video,photo]
colors = ['gold', 'lightskyblue']
explode = (0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

**(b) Word Cloud :**

```python
f = open('bjp_insta.csv', 'r')
df = pd.read_csv(f)
popular = {}
likes=df['likes']
comments=df['comments']
for i in range(0, len(likes)):
    if 'k' in str(likes[i]):
        likes[i] = float(likes[i].strip('k'))*1000
        continue
    likes[i] = ''.join(c for c in likes[i] if c.isnumeric())
    likes[i] = int(likes[i])
for i in range(0, len(comments)):
    if 'k' in str(comments[i]):
        comments[i] = float(comments[i].strip('k'))*1000
        continue
    comments[i] = ''.join(c for c in comments[i] if c.isnumeric())
    comments[i] = int(comments[i])
for i in range(len(df)):
    txt=df['text'][i].strip('<div class="C4VMK"><h2 class="_6lAjh"><a class="FPmhX notranslate TlrDj" href="/bjp4india/" title="bjp4india">bjp4india</a></h2>')
    txt=txt.strip('Linkify">')
    popularity = int(likes[i])*10 + int(views[i])*18 + int(comments[i])*5
    popular[txt]=popularity
sorted_popular = sorted(popular.items(), key=operator.itemgetter(1))
text_words = ' '
for i in range(1,1001):
    s=sorted_popular[len(sorted_popular)-i][0].split()[1:]
    text_words+= ' '.join(s)
```

```python
stopwords = set(STOPWORDS) 
wordcloud = WordCloud(width = 800, height = 800, background_color ='black', stopwords = stopwords, min_font_size = 10).generate(text_words)
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("on") 
plt.tight_layout(pad = 0)   
plt.show()
```

**Most frequently used words :** Narendra Modi, Amit Shah, Vote, BJP, President, Election

**Justification:**

Main motive of this handle is to promote "BJP" party for upcomming elections.

Narendra Modi and Amit Shah are the main faces of the "BJP" party, hence the appear more frequently.

**Note :** Ignoring words like br (break in line for html), href, etc

**(a) Top 5 most popular posts from Instagram's 'explore' feed**

```python
f3 = open('explore_insta.csv', 'r')
df = pd.read_csv(f3)
likes=df['likes']
comments=df['comments']
views=df['views']
popular = {}

changed_likes = []
changed_comments = []
changed_views = []

for i in likes:
    if 'k' in i:
        i=i[:-1]
        i = float(i) * 1000
        i = int(i)

    elif ',' in i:
        i = i.replace(',', '')
        i = int(i)
        
    elif 'm' in i:
        i=i[:-1]
        i = float(i) * 100000
        i = int(i)
    else:
        i = int(i)
        
    changed_likes.append(i)

likes = changed_likes

for i in comments:
    if 'k' in i:
        i=i[:-1]
        i = float(i) * 1000
        i = int(i)

    elif ',' in i:
        i = i.replace(',', '')
        i = int(i)
        
    elif 'm' in i:
        i=i[:-1]
        i = float(i) * 100000
        i = int(i)
    else:
        i = int(i)
        
    changed_comments.append(i)

comments = changed_comments

for i in views:
    if 'k' in str(i):
        i=i[:-1]
        i = float(i) * 1000
        i = int(i)

    elif ',' in str(i):
        i = i.replace(',', '')
        i = int(i)
        
    elif 'm' in str(i):
        i=i[:-1]
        i = float(i) * 100000
        i = int(i)
    else:
        i = int(i)
        
    changed_views.append(i)

views = changed_views
for i in range(len(df)):
    txt = df['id'][i].strip("[]").strip("'")
    popularity = likes[i]*10 + views[i]*18 + comments[i]*5
    popular[txt]=popularity
sorted_popular = sorted(popular.items(), key=operator.itemgetter(1))
for i in range(1,6):
    print(i, "https://www.instagram.com/incindia"+str(sorted_popular[len(sorted_popular)-i][0]))
```
